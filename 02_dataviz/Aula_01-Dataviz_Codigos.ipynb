```python
seno = np.sin(array)
```

```python
plt.figure(figsize=(13, 4))
plt.title('Função seno')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.plot(seno)
plt.plot(seno*-1)

plt.savefig('seno.pdf') # salvar imagem
```

```python
plt.plot(seno, color='green', marker='o')
```

```python
coseno = np.cos(array)
```

```python
plt.plot(seno, label='seno')
plt.plot(coseno, label='coseno')
plt.legend(loc='best')
```

```python
gridsize = (2, 2)

ax1 = plt.subplot2grid(gridsize, loc=(0, 0), colspan=2)
ax2 = plt.subplot2grid(gridsize, loc=(1, 0))
ax3 = plt.subplot2grid(gridsize, loc=(1, 1))

ax1.plot(seno)
ax2.plot(seno*-1)
ax3.plot(coseno)
plt.tight_layout()
plt.show()
```

```python
_, ax = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True)

ax[0, 0].set_title('Seno')
ax[0, 1].set_title('Seno * -1')

ax[0, 0].plot(seno)
ax[0, 1].plot(seno*-1)
ax[1, 0].plot(coseno)
ax[1, 1].plot(coseno*-1)
plt.show()
```
